# Soal 3b

b.Carilah kata-kata slangword yang ada dalam dataset Anda, dengan cara mencocokan dengan kamus KBBI (terlampir).
<br>Simpan hasilnya dalam bentuk csv/excel.
<br> Tampilkan 100 kata slang yang Anda dapatkan dan tampilkan dalam bentuk Gunakan (bobot:25)

```python
#Jawaban
```

```python
data = pd.read_csv('indonesia.csv')
```

```python
kbbi = pd.read_csv('KBBI.csv')
```

```python
slang_words = []
```

```python
for index, row in data.iterrows():
    teks = row['full_text']
    kata_kunci = re.findall(r'\w+', teks)
    for kata in kata_kunci:
        kata = kata.lower()  # Ubah menjadi huruf kecil untuk pencocokan yang tepat
        if kata in kbbi.values:
            slang_words.append(kata)
```

```python
slang_word_counts = pd.Series(slang_words).value_counts().reset_index()
slang_word_counts.columns = ['Kata Slang', 'Jumlah Kemunculan']
slang_word_counts.to_csv('slang_words.csv', index=False)
# Atau untuk menyimpan ke Excel
slang_word_counts.to_excel('slang_words.xlsx', index=False)
```

```python
top_100_slang = slang_word_counts.head(100)
print(top_100_slang)
```
